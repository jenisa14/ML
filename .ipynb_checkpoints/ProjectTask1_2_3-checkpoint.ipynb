```python
empty = df.isnull().sum()

plt.figure(figsize=(10,4))
empty[empty > 0].plot(kind='bar')
plt.title("Missing Values Per Column")
plt.ylabel("Count")
plt.show()


empty

#df.isnull create dataframe of true/false if value is missing->true else 
#.sum -> count true value per column 
# empty ->series of missing values each column
# empty>0 ->filter the coulmns that have missing values
# bar ->shows missing values (x-axis : column names ,y-axis : number of missing values)



```

```python
# handle missing value
# numeric -> median
# categorical -> mode
for col in df.columns:
    if df[col].dtype != 'object':
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna(df[col].mode()[0])

# col->column name
# df[col] ->entire column data
# column->if the column is numerical or not
    # calculate median of column then replace null value with median
    # else -> column is object -> find most frequent value [0] gets the first mode (some colums have multiple modes)
```

```python
# remove outliers
import numpy as np


df_clean = df.copy()

num_cols = df_clean.select_dtypes(include=np.number).columns

for col in num_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # keep only the non-outliers
    df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]

```

```python
# box plot after removing outliers 

num_cols = df_clean.select_dtypes(include=np.number).columns

for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df_clean[col])
    plt.title(f"Boxplot After Outlier Removal - {col}")
    plt.show()
```

```python
df_clean = df.copy()
df_clean
```

```python
# Features (X)
X = df_clean.drop('fraud_reported', axis=1)

# Target (y)
y = df_clean['fraud_reported']

# Categorical columns
cat_cols = X.select_dtypes(include='object').columns

# One-hot encoding
X_encoded = pd.get_dummies(X, columns=cat_cols, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train_enc, X_test_enc, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_enc)
X_test_scaled = scaler.transform(X_test_enc)


```

```python
print(type(X_train_enc))
print(X_train_enc.dtypes.unique())

```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_scaled, y_train)
```

```python
# make prediction
y_pred = lr.predict(X_test_scaled)
```

```python
# fraud probability for each record.
y_prob = lr.predict_proba(X_test_scaled)[:, 1]
y_prob
```

```python
# accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# implement LogisticRegression without using library

# import math library
import math


# Sigmoid function -> Converts any real value into a value between 0 and 1
# Represents probability
# Linear values can be negative or large
# Classification needs probability (0 or 1)
# Sigmoid helps in binary classification

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# Train Logistic Regression function -> Trains the Logistic Regression model
# Finds optimal weight and bias

def train_logistic_regression(X, y, lr, epochs):

    # Epochs define how many times the model iterates over the entire training dataset to minimize error.

    weight = 0
    bias = 0
    n = len(X)

    # Repeats learning process multiple times to reduce error.
    # Increasing epochs helps the model learn better by reducing error.
    
    for _ in range(epochs):
        for i in range(n):

            # Computes weighted sum of input.
            z = weight * X[i] + bias

            # Apply sigmoid -> Converts linear value to probability.
            y_pred = sigmoid(z)

        
            # Gradient calculation
            # calculate error -> Difference between predicted and actual output.
            error = y_pred - y[i]

            # Update weight -> Moves weight in direction of lower error.
            weight -= lr * error * X[i]

            # Update bias -> Adjusts bias to improve accuracy.
            bias -= lr * error

    return weight, bias

# Prediction function -> Uses trained model to make predictions 
# Converts probabilities into class labels (0 or 1)
def predict(X, weight, bias):
    predictions = []
    for x in X:

        # Linear calculation
        z = weight * x + bias
        y_pred = sigmoid(z)
        predictions.append(1 if y_pred >= 0.5 else 0)
    return predictions
```

```python
X = [1, 2, 3, 4, 5]
y = [0, 0, 0, 1, 1]

lr = 0.1
# full reading = 1 epoch
epochs = 1000

weight, bias = train_logistic_regression(X, y, lr, epochs)

print("Weight:", weight)
print("Bias:", bias)

predictions = predict(X, weight, bias)
print("Predictions:", predictions)
```
