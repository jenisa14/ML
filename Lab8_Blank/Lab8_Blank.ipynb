# Do all pre processing on diabetes csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import diabetes.csv data ser using sklearn

```python
df = pd.read_csv("diabetes.csv") 
```

# Check the distribution of the target

```python
df['Outcome'].value_counts()
```

```python
df.columns
```

# Replace zeros with NaN for specific columns

```python
cols_replace_to_zero =['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']
df[cols_replace_to_zero] = df[cols_replace_to_zero].replace(0, np.nan)
```

```python
df[cols_replace_to_zero]
```

# Check for missing values

```python
print(df.isnull().sum())
```

# Fill missing values with median

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

# Convert Targer data into interger code

```python
df['Outcome'] = df['Outcome'].map({
    'Non Diabetic': 0,
    'Diabetic': 1
})
```

```python
sns.heatmap(df.corr(),annot = True,cmap = 'coolwarm')
plt.show()
```
