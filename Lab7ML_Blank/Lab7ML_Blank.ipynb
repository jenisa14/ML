# Decision Tree Classifier

```python
# A Decision Tree Classifier is a supervised ML algorithm used for classification.
# It splits data based on feature values to make decisions.
```

```python
from sklearn.datasets import load_iris 
```

# Import iris data ser using sklearn

```python
data = load_iris()
```

```python
data
```

# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Divide the data into input and output

```python
X = data['data']  #i/p features
```

```python
y = data['target']
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

X_train ,X_test ,y_train ,y_test = train_test_split(

    X, y , test_size = 0.3 , random_state = 42
)
```

# Fitting DecisionTreeClassifier   on dataset

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

# Display Decision Tree

```python
from sklearn import tree
tree.plot_tree(model)
```

# Replace zeros with NaN for specific columns

```python
cols_replace_to_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_replace_to_zero] = df[cols_replace_to_zero].replace(0, np.nan)
```

```python
df[cols_replace_to_zero]
```

# Check for missing values

```python
print(df.isnull().sum())
```

# Fill missing values with median

```python
# perticular one mean
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
# column vise mean
for col in cols_replace_to_zero:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```
