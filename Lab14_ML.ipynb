#  My First Neuron (Perceptron)

**Objective:** Build a single "neuron" from scratch using NumPy to understand how it learns.

**The Logic:**
A Perceptron calculates a weighted sum of inputs. If the sum is big enough (passes a threshold), it fires (Output 1). If not, it stays off (Output 0).

$$ Output = \begin{cases} 1 & \text{if } (w \cdot x + b) \ge 0 \\ 0 & \text{if } (w \cdot x + b) < 0 \end{cases} $$

```python
import numpy as np


def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0

print("Setup complete.")
```

##  Part 1: The OR Gate

 It should output **1** if *any* input is 1.

```python

X_or = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])


y_or = np.array([0, 1, 1, 1])
```

### Step 2: Initialize Weights
We start with random weights because the neuron hasn't learned anything yet.

```python
weights = np.array([0,0]) 
bias = 0
lr = 0.5 

print(f"Initial Weights: {weights}")
print(f"Initial Bias: {bias}")
```

### Step 3: Manual Calculation (Try One Example)
Let's check the math for Input **(0, 1)** manually before automating it.

```python

input_sample = X_or[1]
target_value = y_or[1]


z = np.dot(input_sample, weights) + bias
print(f"Math: ({input_sample[0]} * {weights[0]}) + ({input_sample[1]} * {weights[1]}) + {bias}")
print(f"Sum (z): {z:.2f}")


prediction = step_function(z)
print(f"Prediction: {prediction}")


error = target_value - prediction
print(f"Error: {error}")
```

### Step 4: The Training Loop
We run this for 5 epochs. Watch the "UPDATED" messages to see learning happen.

```python
epochs = 5

print("--- TRAINING ---")

for epoch in range(epochs):
    print(f"\nEPOCH-> {epoch + 1}")
    
    for i in range(4):
        x = X_or[i]
        y = y_or[i]
        
       
        z = np.dot(x, weights) + bias
        pred = step_function(z)
        
    
        error = y - pred
        
       
        if error != 0:
            weights = weights + (lr * error * x)
            bias    = bias    + (lr * error * 1) # always One
            print(f"->MISTAKE on {x}. Expected {y}, got {pred}.")
            print(f"-> UPDATED Weights: {weights}, Bias: {bias:.2f}")
        else:
            print(f"  ✅ Correct on {x}")

print("\nFinal Weights:", weights)
print("Final Bias:", bias)
```
